```python
from google.colab import files
uploaded = files.upload()
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("PS11_Dataset_11.csv")

print("First 5 rows :\n")
print(df.head())
print("\n")

print("Dataset info : \n")
print(df.info())
print("\n")

print("Null Values : \n")
print(df.isnull().sum())
print("\n")

print("Mean :\n")
print(df.mean(numeric_only=True))

print("Median :\n")
print(df.median(numeric_only=True))

print("Stnadar deviation :\n")
print(df.std(numeric_only=True))

numeric_cols = df.select_dtypes(include = np.number).columns
df[numeric_cols].hist(figsize=(12,8))
plt.suptitle("Histogram of Numeric Columns")
plt.tight_layout()
plt.show()

first_numeric = numeric_cols[0]
plt.figure(figsize=(10,6))
plt.plot(df[first_numeric])
plt.title(f"{first_numeric} Trend")
plt.xlabel("Index")
plt.ylabel(first_numeric)
plt.grid(True)
plt.show()

second_numeric = numeric_cols[1]
plt.figure(figsize=(10,6))
plt.plot(df[second_numeric])
plt.title(f"{second_numeric} Trend")
plt.marker = 'o'
plt.xlabel("Index")
plt.ylabel(second_numeric)
plt.grid(True)
plt.show()

third_numeric = numeric_cols[2]
plt.figure(figsize=(10,6))
plt.plot(df[third_numeric])
plt.title(f"{third_numeric} Trend")
plt.xlabel("Index")
plt.ylabel(third_numeric)
plt.grid(True)
plt.show()

fourth_numeric = numeric_cols[3]
plt.figure(figsize=(10,6))
plt.plot(df[fourth_numeric])
plt.title(f"{fourth_numeric} Trend")
plt.xlabel("Index")
plt.ylabel(fourth_numeric)
plt.grid(True)
plt.show()

means = df[numeric_cols].mean()

plt.figure(figsize=(10,5))
plt.bar(means.index, means.values)
plt.title("Average Values")
plt.xticks(rotation=45)
plt.xlabel("Numeric Columns")
plt.ylabel("Mean Values")
plt.tight_layout()
plt.show()

print("\nSummary Insights:")
print("\n2. Statistical analysis completed")
print("\n3. visualization generated")
```
